### 5. Make Table like this to describe the dataframe

#### Columns Explanations :
 - __dataFeatures__ = column name
 - __dataType__ = data type of that column
 - __null__ = how many null data in that column/series
 - __nullPct__ = null percentage of total data in that column
 - __unique__ = how many unique data
 - __uniqueSample__ = get some unique sample from each column

#### hint : search .sample(), .drop_duplicates(), .values, .isna(), dtype

```python
dfTable = pd.DataFrame(columns=[],data=[])
dfTable['dataFeatures'] = list(dfTitanic.columns.values)

dfTable['null'] = list(dfTitanic[list(dfTitanic.columns.values)].isna().sum())

def calcNull(num):
    return "%.2f" % round((num*100)/891 ,2)
dfTable['nullPct'] = list((map(lambda x: calcNull(x),list(dfTable['null']))))

dfTable['unique'] = list(dfTitanic[list(dfTitanic.columns.values)].nunique())

def unique(col):
    return dfTitanic[col].dropna().unique()
dfTable['uniqueSample'] = list((map(lambda x: unique(x),list(dfTable['dataFeatures']))))


dfTable
```

### 8. Missing Values
#### Berapa banyak NaN age data dan mean dari setiap who?
#### Isi data NaN dari kolom age dengan mean dari umur tergantung whonya dia apa (child/woman/man)
#### Kemudian print banyak NaN age data dan mean umur setiap who lagi.

```python
who = ['man', 'woman', 'child']
for i in range(len(who)):
    df = dfTitanic[dfTitanic['who'] == who[i]]['age']
    nanVal = df.isna().sum()
    meanVal = df.mean()
    print('Banyak Nan ' + who[i] + ' & mean: ' + str(nanVal) + ' & ' + str(meanVal))
```

```python
for i in range(len(dfTitanic)):
    if dfTitanic['who'][i] == 'man':
        if dfTitanic['age'][i] > 0:
            dfTitanic['age'][i] = dfTitanic['age'][i]
        else:
            dfTitanic['age'][i] = 33.17312348668281
    elif dfTitanic['who'][i] == 'woman':
        if dfTitanic['age'][i] > 0:
            dfTitanic['age'][i] = dfTitanic['age'][i]
        else:
            dfTitanic['age'][i] = 32.0
            
who = ['man', 'woman', 'child']
for i in range(len(who)):
    df = dfTitanic[dfTitanic['who'] == who[i]]['age']
    nanVal = df.isna().sum()
    meanVal = df.mean()
    print('Banyak Nan ' + who[i] + ' & mean: ' + str(nanVal) + ' & ' + str(meanVal))
```

### 9. Adding New Features/Columns/Variables

#### Buat DataFrame baru dengan column2 sebagai berikut
 - __id__ = id table ini (PK)
 - __mean_fare__ = Mean Fare
 - __std_fare__ = Std Fare
 - __batas_min_fare__ = Mean Fare - (3 x Std Fare)
 - __batas_max_fare__ = Mean Fare + (3 x Std Fare)
 
#### Dan data2nya ada 3 (fare setiap class yaitu first, second, and third class)

```python
cate = [3,1,2]
dfStd = pd.DataFrame(columns=['id', 'mean_fare', 'std_fare', 'batas_min_fare', 'batas_max_fare'],
                       data=[
                           [i+1,
                            dfTitanic[dfTitanic['pclass'] == cate[i]]['fare'].mean(),
                            dfTitanic[dfTitanic['pclass'] == cate[i]]['fare'].std(),
                            dfTitanic[dfTitanic['pclass'] == cate[i]]['fare'].mean()-(3*dfTitanic[dfTitanic['pclass'] == cate[i]]['fare'].std()),
                            dfTitanic[dfTitanic['pclass'] == cate[i]]['fare'].mean()+(3*dfTitanic[dfTitanic['pclass'] == cate[i]]['fare'].std())
                           ] for i in range(len(cate)) ])
dfStd
```

#### Print outlier % of each class

```python
first = list(dfTitanic[dfTitanic['pclass'] == 1]['fare'])

fCount = 0
for i in range(len(first)):
    if ((first[i] > dfStd['batas_max_fare'][1])  | (first[i] < dfStd['batas_min_fare'][1])):
        fCount += 1

second = list(dfTitanic[dfTitanic['pclass'] == 2]['fare'])

sCount = 0
for i in range(len(second)):
    if ((second[i] > dfStd['batas_max_fare'][2])  | (second[i] < dfStd['batas_min_fare'][2])):
        sCount += 1
        
third = list(dfTitanic[dfTitanic['pclass'] == 3]['fare'])

tCount = 0
for i in range(len(third)):
    if ((third[i] > dfStd['batas_max_fare'][0])  | (third[i] < dfStd['batas_min_fare'][0])):
        tCount += 1

print('Outlier % First : ' + str((fCount * 100)/ len(first)))
print('Outlier % Second : ' + str((sCount * 100)/ len(second)))
print('Outlier % Third : ' + str((tCount * 100)/ len(third)))
```

#### Tambahkan column outliercalcId (FK) di dfTitanic

```python
def outlier(x):
    if x == 1:
        return 2
    elif x == 2:
        return 3
    else:
        return 1
dfTitanic['outliercalcId'] = list(map(lambda x:outlier(x), list(dfTitanic['pclass'])))
```

```python
dfTitanic.head()
```

#### Tambahkan column outlier (1 for yes and 0 for no)

```python
yesNo = []
for i in range(len(dfTitanic)):
    if dfTitanic['pclass'][i] == 1:
        if ((dfTitanic['fare'][i] > dfStd['batas_max_fare'][1])  | (dfTitanic['fare'][i] < dfStd['batas_min_fare'][1])):
            yesNo.append(1)
        else:
            yesNo.append(0)
        
    if dfTitanic['pclass'][i] == 2:
        if ((dfTitanic['fare'][i] > dfStd['batas_max_fare'][2])  | (dfTitanic['fare'][i] < dfStd['batas_min_fare'][2])):
            yesNo.append(1)
        else:
            yesNo.append(0)
            
    if dfTitanic['pclass'][i] == 3:
        if ((dfTitanic['fare'][i] > dfStd['batas_max_fare'][0])  | (dfTitanic['fare'][i] < dfStd['batas_min_fare'][0])):
            yesNo.append(1)
        else:
            yesNo.append(0)
            
dfTitanic['outlier'] = yesNo
dfTitanic.head()
```
